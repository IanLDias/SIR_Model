```python
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
def SIR(b,v, S_0, I_0, time, lockdown=False):
    S = S_0
    I = I_0
    R = [0, ]
    for i in range(time):
        S.append(S[i]+(-b*I[i]*S[i]))
        I.append(I[i]+(b*I[i]*S[i] - v*I[i]))
        R.append(R[i]+(v*I[i]))

    fig = go.Figure()
    fig.add_trace(go.Scatter(y=S, text='Susceptible'))
    fig.add_trace(go.Scatter(y=I, text='Infected'))
    fig.add_trace(go.Scatter(y=R, text='Recovered'))
    return fig
```

```python
SIR(b=0.7/500,v=1/5,S_0=[499,],I_0=[5,],time=100)
```
